## Univariate analysis and visualization

```python

sns.set_style("whitegrid")   # clean background
plt.figure(figsize=(15, 10))

colors = sns.color_palette("viridis", len(numerical_cols))  # nice color palette

for i, col in enumerate(numerical_cols):
    plt.subplot(4, 3, i + 1)
    sns.histplot(df[col], kde=True, color=colors[i], bins=30)
    plt.title(col, fontsize=12, fontweight="bold")
    plt.xlabel(col, fontsize=10)
    plt.ylabel("Count")

plt.tight_layout()
plt.show()
```

## Select numerical columns for correlation analysis

```python

numerical_cols = ['Age', 'Estimated Income', 'Superannuation Savings', 'Credit Card Balance',
                  'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts',
                  'Foreign Currency Account', 'Business Lending', 'Properties Owned']

# Calculate the correlation matrix
correlation_matrix = df[numerical_cols].corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='YlGnBu', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()

```

## Bivariate Analysis: Regression Plots of Selected Feature Pairs


```python
pairs_to_plot = [
    ('Bank Deposits', 'Saving Accounts'),
    ('Checking Accounts', 'Saving Accounts'),
    ('Checking Accounts', 'Foreign Currency Account'),
    ('Age', 'Superannuation Savings'),
    ('Estimated Income', 'Checking Accounts'),
    ('Bank Loans', 'Credit Card Balance'),
    ('Business Lending', 'Bank Loans'),
]


for x_col, y_col in pairs_to_plot:
    plt.figure(figsize=(8, 6))
    sns.regplot(
        data=df,
        x=x_col,
        y=y_col,
        scatter_kws={'alpha': 0.4},     # semi-transparent points
        line_kws={'color': 'red'}       # best-fit line color
    )
    plt.title(f'Relationship between {x_col} and {y_col}', fontsize=14)
    plt.xlabel(x_col, fontsize=12)
    plt.ylabel(y_col, fontsize=12)
    plt.tight_layout()
    plt.show()
```

# Insights:


1. **Deposits and Savings Behavior**  
   The strong correlation between *Bank Deposits* and *Saving Accounts* indicates overlapping financial behavior: customers who frequently deposit funds are also more likely to maintain or grow their savings balances. This suggests these features may capture similar aspects of liquidity preference.

2. **Income, Age, and Wealth Accumulation**  
   Moderate correlations of *Age* and *Estimated Income* with *Superannuation*, *Savings*, and *Checking Accounts* reflect a financial lifecycle pattern. Older and higher-income individuals typically accumulate greater savings and retirement funds, but may also hold higher credit card balances and loans.

3. **Low Correlation with Property Ownership**  
   The weak relationship between *Properties Owned* and other banking variables implies that property acquisition depends on external factors (e.g., real estate markets, inheritance, location) rather than core banking activity.

4. **Business vs. Personal Banking**  
   *Business Lending* shows moderate correlation with *Bank Loans*, suggesting some overlap between personal and business borrowing. However, its low correlation with deposits and property-related metrics indicates it largely serves a distinct customer segment with separate financial behaviors.
